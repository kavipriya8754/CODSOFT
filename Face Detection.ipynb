```python
import cv2

import os

# Your file path
cascade_path ="C:\Face Recognition\Cascades\\fullbody.xml"

# Check if the file exists
if not os.path.exists(cascade_path):
    print(f"Error: Cascade file not found at {cascade_path}")
else:
    # Load the cascade classifier
    full_detector = cv2.CascadeClassifier(cascade_path)

    # Check if the cascade classifier is loaded successfully
    if full_detector.empty():
        print("Error: Cascade classifier not loaded")
    else:
    # Read the image
        image = cv2.imread("C:\Face Recognition\Data Sets\people3.jpg")

    # Convert the image to grayscale
        image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect objects in the image
        detections = full_detector.detectMultiScale(image_gray, scaleFactor=1.03, minNeighbors=1)

    # Draw rectangles around the detected objects
        for (x, y, w, h) in detections:
            cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Display the result
            cv2.imshow('Object Detection', image)
            cv2.waitKey(1000)
            cv2.destroyAllWindows()
```
